```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,2,3],[40,59,67])
plt.title("Yash's Data")
plt.xlabel("This is x")
plt.ylabel("This is y")
plt.show()
```

```python
x = [1,5,7]
y = [23,56,75]
x1 = [5,6,7]
y1 = [9,7,6]
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

```python
from matplotlib import style
```

```python
style.use('ggplot')
x = [1,5,7]
y = [23,56,75]
x1 = [5,6,7]
y1 = [9,7,6]
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

```python
style.use('ggplot')
x = [1,5,7]
y = [23,56,75]
x1 = [5,6,7]
y1 = [9,7,6]
plt.plot(x,y,label="first")
plt.plot(x1,y1,label="second")
plt.legend()
plt.show()
```

```python
style.use('ggplot')
x = [1,5,7]
y = [23,56,75]
x1 = [5,6,7]
y1 = [9,7,6]
plt.plot(x,y,label="first",linewidth=6)
plt.plot(x1,y1,label="second",linewidth=1)
plt.legend()
plt.show()
```

```python
style.use('ggplot')
x = [1,5,7]
y = [23,56,75]
x1 = [5,6,7]
y1 = [9,7,6]
plt.scatter(x,y,label="first")
plt.hist(x1,label="second")
plt.legend()
plt.show()
```
