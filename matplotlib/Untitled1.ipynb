```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pencil_price=[1,2,3,4,5,6,7,8,9]
eraser_price=[9,8,7,6,5,4,3,2,1]
year=[2000,2002,2004,2006,2008,2010,2012,2014,2016]
```

```python
plt.plot(year,pencil_price)
plt.show()
```

```python
plt.plot(year,eraser_price)
plt.show()
```

```python
plt.xlabel('year')
plt.ylabel('price')
```

```python
plt.scatter(year,pencil_price)
```

```python
plt.plot(year,pencil_price,
         year,eraser_price)
```

```python
plt.plot(year,pencil_price,'--y',
         year,eraser_price,':c')
```

```python
plt.axis([1900,2100,0,20])
plt.show()
```
